Firstly, we import the necessary libraries

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
```

Now, we import the dataset as a dataframe. But we need to perform some pre-processing on the data.    
The different 'activities' involved in the data are qualitative. We need to quantify the activities in order to train the models. So, we assign each of the activity with a number. 

```python
data = pd.read_csv('HAR dataset.csv')
data = data.dropna()
data['Activity_no'] = pd.factorize(data['Activity'])[0]
data.head()
```

```python
data['Activity'].unique()
```

```python
data['Activity'].value_counts()
```

Now to prepare the dataset for training our model, we separate the activity label from the input variables, and store the input variables as X. 

```python
X = data.columns.to_list()
X.remove('Activity')
X.remove('Activity_no')
X = data[X]
```

```python
X
```

We are ready to train our data now. But before that, we will split our data into train and test set. 

```python
X_train, X_test, y_train, y_test = train_test_split(X, data['Activity_no'], test_size=0.2, random_state=42)
```

Firstly, a Random Forest Classifier model is trained on the training dataset and its accuracy is estimated using the accuracy score function on the test data. 

```python
model_rfc = RandomForestClassifier()
model_rfc.fit(X_train, y_train)
```

```python
y_pred_model_rf = model_rfc.predict(X_test)
print(accuracy_score(y_test, y_pred_model_rf))
```

Finally, we train a SVC model on the train data and find its accuracy on the test data.

```python
model_svc = SVC(kernel = 'linear')
model_svc.fit(X_train, y_train)
```

```python
y_pred_model_svc = model_svc.predict(X_test)
print(accuracy_score(y_test, y_pred_model_svc))
```

It is found that for this problem, the SVC performs better than the Random Forest Classifier based on the test scores.
